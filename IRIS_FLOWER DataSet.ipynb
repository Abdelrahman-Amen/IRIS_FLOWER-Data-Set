```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("C:\\Users\\abdel\\Desktop\\\iris\\irisdataset.csv",names=["sepal_length","sepal_width",
"petal_length","petal_width","species"])
data=shuffle(data)
print(data.head(10))                                                                        
```

```python
Setosa=data[data["species"].isin(["Iris-setosa"])]
Versicolour=data[data["species"].isin(["Iris-versicolor"])]
Virginica=data[data["species"].isin(["Iris-virginica"])]
```

```python
fig,ax=plt.subplots(figsize=(10,10))
ax.scatter(Setosa["sepal_length"],Setosa["sepal_width"],c="b",marker="o",label="Setosa")
ax.scatter(Versicolour["sepal_length"],Versicolour["sepal_width"],c="g",marker="o",label="Versicolour")
ax.scatter(Virginica["sepal_length"],Virginica["sepal_width"],c="orange",marker="o",label="Virginica")
leg=ax.legend()
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
ax.set_title("classification of Setosa,Versicolour and Virginica according to sepal_length and sepal_width")
```

```python
fig,ax=plt.subplots(figsize=(10,10))
ax.scatter(Setosa["petal_length"],Setosa["petal_width"],c="b",marker="o",label="Setosa")
ax.scatter(Versicolour["petal_length"],Versicolour["petal_width"],c="orange",marker="o",label="Versicolour")
ax.scatter(Virginica["petal_length"],Virginica["petal_width"],c="g",marker="o",label="Virginica")
leg=ax.legend()
ax.set_xlabel("petal_length")
ax.set_ylabel("petal_width")
ax.set_title("classification of Setosa,Versicolour and Virginica according to petal_length and petal_width")
```
